# COMBINACIÓN DE ESTRUCTURAS DE CONTROL

```python
i=1
while i<6:
    print(i)
    if i ==3:
       break
    i+=1

i=1
while i<6:
    i+=1
    if i==3:
        continue
    print(i)

i=1
while i<6:
    print(i)
    i+=1
else:
    print("i ya no es menor que 6")

for x in range (5,30,3):
    print(x)

for x in range (6):
    print(x)
else: 
    print("!Terminó!")

característica=["roja","grande","rica"]
fruta=["manzana","uva","fresa"]
for x in fruta:
    for y in característica:
        print(x,y)

x=int(input("Ingrese un número\n"))
y=int(input("Ingrese un número\n"))
if x>y:
    print(x,"es mayor que",y)
else: 
    print(y,"es mayor que",x)

for x in range(20):
    if (x>4 and x<10):
        print(x)

for x in range (20):
    if (x<4 and x>10):
        print(x)
```
